## Different methods of getting data

```python
import requests
import json 
import pandas as pd
```

```python
# stocks
url = "http://quotes.sina.cn/cn/api/json_v2.php/CN_MarketDataService.getKLineData?symbol=sh000001&scale=15&datalen=1023"

req = requests.get(url)
results = json.loads(req.content)
df = pd.DataFrame(results).set_index("day")
```

```python
df
```

```python
# futures
url = "https://quotes.sina.cn/futures/api/json.php/CffexFuturesService.getCffexFuturesMiniKLine5m?symbol=IF1908"
req = requests.get(url)
results = json.loads(req.content)

l = []
for line in results:
    bar = {}
    bar["date"] = line[0]
    bar["open"] = float(line[1])
    bar["high"] = float(line[2])
    bar["low"] = float(line[3])
    bar["close"] = float(line[4])
    bar["vol"] = int(line[5])
    l.append(bar)
    
df = pd.DataFrame(l).set_index("date")
df
```
